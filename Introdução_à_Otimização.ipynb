## Introdução à Otimização

Otimização é o processo de tornar algo o melhor possível ou mais eficiente.

Exemplos de otimização:

* Encontrar o caminho mais curto de um ponto A até um ponto B.
* Aumentar o lucro de um determinado negócio.
* Reduzir o tempo gasto para concluir uma tarefa.
* Encontrar a melhor carteira de investimento.
* Outros.



---



### Variáveis de Decisão

Para tornar algo o melhor possível ou mais eficiente, normalmente temos algumas escolhas a serem feitas.

Essas escolhas são representadas por **variáveis de decisão**.

As variáveis de decisão podem assumir diferentes tipos numéricos, dependendo da natureza do problema. Aqui estão os tipos comuns de variáveis de decisão:

* Variáveis Contínuas: as variáveis podem assumir qualquer valor real em um intervalo específico. Exemplo: Valor, em Reais, que posso gastar em compra de ativos.

* Variáveis Inteiras: as variáveis podem assumir qualquer valor inteiro em um intervalo específico. Exemplo: Quantidade de papeis ITUB4 que vou comprar.

* Variáveis Binárias: as variáveis assumem apenas dois valores, geralmente 0 ou 1. Exemplo: Comum onde a decisão é do tipo "sim" ou "não", como decidir se usar ou não um recurso.



É importante destacar que as variáveis de decisão são variaveis que são controladas por nós. Se eu não possuo o controle sobre aquela variável, eu não devo utilizar ela em um processo de otimização porque eu não vou conseguir garantir a aplicação do valor encontrado.

Exemplo:
Considerar o preço do ITUB4 como uma variável de decisão. Encontro no processo de otimização que o preço do ativo tem que ser igual o preço dos seus dividendos para que eu possa gerar dinheiro infinito.

### Função Objetivo

Em qualquer problema de otimização, há algo que você deseja maximizar ou minimizar.

Relembrando dois exemplos de otimização:

* Aumentar o lucro de um determinado negócio. Neste caso, desejamos maximizar.
* Reduzir o tempo gasto para concluir uma tarefa. Neste caso, desejamos minimizar.

Estes "objetivos" podem ser representados através de funções que informam o quanto estamos máximizando ou minimizando o que desejamos. Estas funções podem ser lineares ou não lineares, dependendo da complexidade do problema.

Vamos a um exemplo prático:

Quero construir uma função objetivo para representar o aumento do lucro do meu negócio a cada item que eu vendo. Cada item vale 10 reais. Considerando a quantidade de itens vendidos como uma variável de decisão, temos:

$f(vendidos) = 10 * vendidos$

```python
def f(vendidos):
  return 10 * vendidos
```

```python
# Se eu vender 5 itens
vendidos = 5
lucro = f(vendidos)
print(f"Ao vender {vendidos}, tive um lucro de R${lucro}.")
```

```python
# Se eu vender 5.5 itens (kilos de um produto)
vendidos = 5.5
lucro = f(vendidos)
print(f"Ao vender {vendidos}, tive um lucro de R${lucro}.")
```

```python
# Se eu vender -10 itens
vendidos = -10
lucro = f(vendidos)
print(f"Ao vender {vendidos}, tive um lucro de R${lucro}.")
```

```python
# @title
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Configurando o estilo do seaborn
sns.set(style="whitegrid")

# Gerando valores no intervalo de -10 a 100
x_values = np.arange(-10, 101, 1)

# Calculando os valores correspondentes a função objetivo
y_values = f(x_values)

# Criando o gráfico usando seaborn
plt.figure(figsize=(8, 6))
sns.lineplot(x=x_values, y=y_values)

# Adicionando rótulos e título
plt.xlabel('Vendidos')
plt.ylabel('f(vendidos)')

# Exibindo o gráfico
plt.show()
```

$g(vendidos) = 10 * vendidos - 0.05 * vendidos^2$

```python
def g(vendidos):
  return 10 * vendidos - 0.05 * vendidos * vendidos
```

Em alguns casos, podemos ter mais de uma função objetivo que desejamos máximizar ou minimizar.

Normalmente estas funções tendem a ser conflitantes, ou seja, a melhoria em uma leva a um resultado pior na outra.

Para exemplificar, suponha que quero reduz o meu tempo trabalhando mas, ao mesmo tempo, ainda quero maximizar meu lucro.

Se eu gasto 2 horas produzindo um item que vou vender, posso expressão uma função objetivo de tempo trabalhado como:
$h(vendidos) = 2 * vendidos$

Quanto mais eu vendo, maior meu lucro e meu tempo de trabalho. Mas se eu quero trabalhar menos, quando menos eu produzo, menor meu lucro.

Existem geralmente limitações ou restrições em nossas escolhas.

Isso pode ser o tempo disponível, o dinheiro que você pode gastar, ou qualquer outro fator que influencie suas decisões.

Por exemplo:

Apesar de eu poder vender quantos itens eu tiver à minha disposição, eu não consigo vender um número negativo de itens.

Deste modo, devemos ter para este caso um limite inferior para a quantidade de vendidos.

Ela pode ser escrita como: $vendidos \geq 0$.


Outro Exemplo:

Quero comprar ITUB4, mas só posso gastar R$ 5000. Se o valor gasto é uma variável de decisão, temos as seguintes restrições:


*   Limite Inferior: $gastos \geq 0$
*   Limite Superior: $gastos \leq 5000$

As restrições que envolve limites inferiores e superiores em variáveis são comumente chamado de "restrição de limite" ou "restrição de intervalo".

Essas restrições são usadas para impor limites específicos às variáveis de decisão em um problema de otimização.

```python
# Gerando valores no intervalo de -10 a 100
x_values_1 = np.arange(-10, 101, 1)

# Gerando valores no intervalo valido
x_values_2 = np.arange(0, 51, 1)

# Calculando os valores correspondentes a função objetivo
y_values_1 = f(x_values_1)
y_values_2 = f(x_values_2)

# Criando o gráfico usando seaborn
plt.figure(figsize=(8, 6))
sns.lineplot(x=x_values_1, y=y_values_1)
sns.lineplot(x=x_values_2, y=y_values_2, color="red")


# Adicionando rótulos e título
plt.xlabel('Vendidos')
plt.ylabel('f(vendidos)')

# Exibindo o gráfico
plt.show()
```

### Restrições de Desigualdade e Igualdade

Da mesma forma com as restrições de limites, as restrições de desigualdade e igualdade são fundamentais para definem as limitações e condições sob as quais a função objetivo deve ser otimizada.

Vamos exemplificar:

Suponha que temos agora dois itens para ser vendidos, o item A e o item B. Eu ganho 10 reais se vender o produto A e ganho 8 reais se vender o produto B.

Apesar disso, para produzir estes produtos eu preciso de madeira e prego.

Para o produto A eu preciso de 2 madeiras e 4 pregos, enquanto para o produto B eu preciso de 3 madeiras e 1 pregos.

Como os recursos não são ilimitados, eu tenho apenas 40 unidades de madeira e 25 unidades de prego.

Analisando eu tenho exatamente duas restrições de desigualdade aqui, uma para a quantidade de madeiras que eu tenho e outra para quantidade de pregos. São elas:


*   Madeira: $2A + 3B \leq 40$
*   Pregos: $4A + 1B \leq 25$





```python
# Definindo as restrições
x = np.linspace(0, 20, 400)
y1 = (40 - 2*x) / 3
y2 = (25 - 4*x) / 1

# Criando o gráfico
plt.figure(figsize=(8, 6))

# Adicionando as regiões sombreadas
plt.fill_between(x, 0, y1, where=(y1 >= 0), color='skyblue', alpha=0.4, label=r'$2A + 3B \leq 40$')
plt.fill_between(x, 0, y2, where=(y2 >= 0), color='lightcoral', alpha=0.4, label=r'$4A + 1B \leq 25$')

# Adicionando as linhas das restrições
plt.plot(x, y1, label=r'$2A + 3B \leq 40$', color='blue', linestyle='dashed')
plt.plot(x, y2, label=r'$4A + 1B \leq 25$', color='red', linestyle='dashed')

# Adicionando as linhas x >= 0 e y >= 0
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Ajustando o layout
plt.xlim(0, 15)
plt.ylim(0, 15)
plt.xlabel(r'$A$')
plt.ylabel(r'$B$')

# Adicionando legendas
plt.legend()

# Exibindo o gráfico
plt.show()
```

Se além disso, eu ainda quiser incluir uma restrição de tempo de trabalho, onde eu quero trabalhar exatamente 40 horas e depois abandonar o problema.

Se eu gasto 5 horas para fazer o produto A e 6 horas para o produto B, minha restrição de tempo ficaria:

$5A + 6B = 40$

```python
# Definindo as restrições
x = np.linspace(0, 20, 400)
y1 = (40 - 2*x) / 3
y2 = (25 - 4*x) / 1
y3 = (40 - 5*x) / 6  # Nova restrição de igualdade

# Criando o gráfico
plt.figure(figsize=(8, 6))

# Adicionando as regiões sombreadas
plt.fill_between(x, 0, y1, where=(y1 >= 0), color='skyblue', alpha=0.4, label=r'$2A + 3B \leq 40$')
plt.fill_between(x, 0, y2, where=(y2 >= 0), color='lightcoral', alpha=0.4, label=r'$4A + 1B \leq 25$')

# Adicionando as linhas das restrições
plt.plot(x, y1, label=r'$2A + 3B \leq 40$', color='blue', linestyle='dashed')
plt.plot(x, y2, label=r'$4A + 1B \leq 25$', color='red', linestyle='dashed')
plt.plot(x, y3, label=r'$5A + 6B = 40$', color='green', linestyle='dashed')

# Adicionando as linhas x >= 0 e y >= 0
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Ajustando o layout
plt.xlim(0, 15)
plt.ylim(0, 15)
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')

# Adicionando legendas
plt.legend()

# Exibindo o gráfico
plt.show()
```

## Otimização Quadrática

A otimização quadrática refere-se à maximização ou minimização de uma função quadrática.

Uma função quadrática é uma função polinomial de segundo grau e tem a forma geral, para uma única variavel, igual a:

$f(x) =ax^2 + bx + c$.

Para duas variaveis:

$f(x, y) = ax^2 + bxy + cy^2 + dx + ey + f$

E assim por diante...

Para resolver um problema do tipo quadrático podemos seguir os seguintes passos:

1. Encontre o ponto crítico: Calcule a derivada da função em relação a x. Em seguida, iguale a derivada a zero e encontre os valores de x que vão satisfazer esta esta condiçao.
2. Verifique a concavidade: Use a segunda derivada para verificar se o ponto crítico é um mínimo ou máximo. Se a segunda derivada for positiva, é um mínimo; se for negativa, é um máximo.
3. Encontre o valor correspondente de y: Substitua o valor de x encontrado nos passos anteriores na função original para encontrar o valor mínimo (ou máximo) da função quadrática.

Vamos para um exemplo:

Considerando a função $g(vendidos) = 10 * vendidos - 0.05 * vendidos^2$, onde queremos maximizar o lucro.

```python
# Gerando valores no intervalo de -10 a 130
x_values = np.arange(-10, 131, 1)

# Calculando os valores correspondentes a função objetivo
y_values = g(x_values)

# Criando o gráfico usando seaborn
plt.figure(figsize=(8, 6))
sns.lineplot(x=x_values, y=y_values)

# Adicionando rótulos e título
plt.xlabel('Vendidos')
plt.ylabel('g(vendidos)')

# Exibindo o gráfico
plt.show()
```

```python
from sympy import diff, symbols, Eq, solve

x = symbols('vendidos')
g = 10 * x - 0.05 * x*x
print(g)
```

```python
# derivada primeira
der_pri_g = diff(g, x, 1)
print(der_pri_g)
```

```python
# Crie uma equação definindo a expressão igual a zero
equacao = Eq(der_pri_g, 0)

# Resolva a equação para encontrar os valores de x que satisfazem a condição
solucoes = solve(equacao, x)

print(solucoes)
```

```python
# derivada segunda
der_seg_g = diff(g, x, 2)
print(der_seg_g)
```

Se a segunda derivada for positiva, é um mínimo; se for negativa, é um máximo.

No caso, encontramos que é um valor negativo, igual a $-0.10$.

Como encontramos apenas um ponto, o $vendidos = 100$, sabemos que ele é um ponto de máximo.

Substituindo este valor na equação temos o nosso lucro máximo:

```python
# Avalie as soluções na expressão original
for solucao in solucoes:
    resultado = g.subs(x, solucao)
    print(f"Para vendidos = {solucao}, o lucro é de {resultado}.")
```

Bem legal né? E se o problema fosse um pouco mais complexo, com duas variáveis?

Seja então um problema de minimização da função:

$f(x,y)= 2x^2 + 3y^2 + 2xy + 4x + 5y + 6$



```python
x, y = symbols('x y')

f = 2*x**2 + 3*y**2 + 2*x*y + 4*x + 5*y + 6
print(f)
```

```python
der_pri_f_x = diff(f, x, 1)
der_pri_f_y = diff(f, y, 1)

print("Derivada parcial em x: ", der_pri_f_x)
print("Derivada parcial em y: ", der_pri_f_y)
```

```python
# Encontrar os pontos críticos

# Crie as equações igualadas a zero
eq1 = Eq(der_pri_f_x, 0)
eq2 = Eq(der_pri_f_y, 0)

# Resolva o sistema de equações
solution = solve((eq1, eq2), (x, y))

# Exiba os resultados
print("Pontos críticos (x, y):", solution)
```

Para determinar a concavidade, precisamos da derivada segunda. Desta forma, precisamos da matriz Hessiana, conforme:


\begin{bmatrix}
  \frac{\partial^2 f}{\partial x^2} & \frac{\partial^2 f}{\partial x \partial y} \\
  \frac{\partial^2 f}{\partial y \partial x} & \frac{\partial^2 f}{\partial y^2}
\end{bmatrix}

```python
# Verifique a concavidade

from sympy import hessian

# Crie a matriz Hessiana
hessian_matrix = hessian(f, (x, y))

print("Matriz Hessiana:")
print(hessian_matrix)
```

E como verifico se é um ponto de mínimo, máximo ou de sela?

* Se todos os auto-valores da Hessiana no ponto crítico são positivos, f tem um mínimo relativo no ponto analisado.
* Se todos os auto-valores da Hessiana no ponto crítico são negativos, f tem um máximo relativo no ponto analisado.
* Se a Hessiana no ponto crítico tem auto-valores positivos e negativos, o ponto analisado é um ponto de sela de f.

```python
# Calcule os autovalores
autovalores = hessian_matrix.eigenvals()

# Exiba os resultados
print("Autovalores:", autovalores)
```

```python
# Avalie a função em pontos específicos
resultado = f.subs(solution)
print(f"f({solution}) = {resultado}")
```

### Penalização

Restrição de Igualdade:

Se você tem uma restrição do tipo Ax=b, onde A é uma matriz, x é o vetor de variáveis de decisão e b é um vetor constante, você pode penalizar a diferença entre o lado esquerdo e o lado direito da equação quadrada.

A penalização pode ser modelada da seguinte forma:

$Penalização = \lambda ||Ax -b||^2$

λ é um parâmetro que controla o peso da penalização na função objetivo.

O valor da penalização deve ser somada ao valor da função objetivo.

Vamos supor que temos que minimizar a função seno, com $1 \leq x \leq 15$.

```python
# Gerando valores no intervalo de -10 a 130
x_values = np.arange(1, 15, 0.001)

# Calculando os valores correspondentes a função objetivo
y_values = np.sin(x_values)

# Criando o gráfico usando seaborn
plt.figure(figsize=(8, 6))
sns.lineplot(x=x_values, y=y_values)

# Adicionando rótulos e título
plt.xlabel('Vendidos')
plt.ylabel('g(vendidos)')

# Exibindo o gráfico
plt.show()
```

Só que temos uma restrição de igualdade:

$10 x = 50$

$x = 5$

Se $\lambda=0.0001$, temos:

```python
lambda_value = 0.0001

# Gerando valores no intervalo de -10 a 130
x_values = np.arange(1, 15, 0.001)

# Calculando os valores correspondentes a função objetivo
y_values = np.sin(x_values) + lambda_value * (10*x_values - 50)**2

# Criando o gráfico usando seaborn
plt.figure(figsize=(8, 6))
sns.lineplot(x=x_values, y=y_values)

# Adicionando rótulos e título
plt.xlabel('Vendidos')
plt.ylabel('g(vendidos)')

# Exibindo o gráfico
plt.show()
```

Se $\lambda=0.001$, temos:

```python
lambda_value = 0.001

# Gerando valores no intervalo de -10 a 130
x_values = np.arange(1, 15, 0.001)

# Calculando os valores correspondentes a função objetivo
y_values = np.sin(x_values) + lambda_value * (10*x_values - 50)**2

# Criando o gráfico usando seaborn
plt.figure(figsize=(8, 6))
sns.lineplot(x=x_values, y=y_values)

# Adicionando rótulos e título
plt.xlabel('Vendidos')
plt.ylabel('g(vendidos)')

# Exibindo o gráfico
plt.show()
```

Se $\lambda=1$, temos:

```python
lambda_value = 1

# Gerando valores no intervalo de -10 a 130
x_values = np.arange(1, 15, 0.001)

# Calculando os valores correspondentes a função objetivo
y_values = np.sin(x_values) + lambda_value * (10*x_values - 50)**2

# Criando o gráfico usando seaborn
plt.figure(figsize=(8, 6))
sns.lineplot(x=x_values, y=y_values)

# Adicionando rótulos e título
plt.xlabel('Vendidos')
plt.ylabel('g(vendidos)')

# Exibindo o gráfico
plt.show()
```

Restrição de Desigualdade:

Se você tem uma restrição do tipo Ax≤b, onde A é uma matriz, x é o vetor de variáveis de decisão e b é um vetor constante, você pode penalizar as violações da desigualdade.

A penalização pode ser modelada da seguinte forma:

$Penalização = \lambda . max(0, Ax - b)$

λ é um parâmetro que controla o peso da penalização na função objetivo.

O valor da penalização deve ser somada ao valor da função objetivo.

Vamos supor, novamente, que temos que minimizar a função seno, com $1 \leq x \leq 15$.

```python
# Gerando valores no intervalo de -10 a 130
x_values = np.arange(1, 15, 0.001)

# Calculando os valores correspondentes a função objetivo
y_values = np.sin(x_values)

# Criando o gráfico usando seaborn
plt.figure(figsize=(8, 6))
sns.lineplot(x=x_values, y=y_values)

# Adicionando rótulos e título
plt.xlabel('Vendidos')
plt.ylabel('g(vendidos)')

# Exibindo o gráfico
plt.show()
```

Só que temos uma restrição de desigualdade:

$5 x \leq 50$

$x \leq 10$

Se $\lambda=0.1$, temos:

```python
lambda_value = 0.1

# Gerando valores no intervalo de -10 a 130
x_values = np.arange(1, 15, 0.001)

# Calculando os valores correspondentes a função objetivo
y_values = np.sin(x_values) + lambda_value * np.maximum(0, 5*x_values - 50)

# Criando o gráfico usando seaborn
plt.figure(figsize=(8, 6))
sns.lineplot(x=x_values, y=y_values)

# Adicionando rótulos e título
plt.xlabel('Vendidos')
plt.ylabel('g(vendidos)')

# Exibindo o gráfico
plt.show()
```

## Otimização Binária Quadrática Irrestrita
### (QUBO - Quadratic Unconstrained Binary Optimization)

Ao chegar aqui, já conhecemos bastante coisa. Se eu falo que quero trabalhar em um problema QUBO, que em portugues significa Otimização Binária Quadrática Irrestria, eu sei que:

* A função objetivo é quadrática.
  * É uma função polinomial de segundo grau.
* O problema não possui restrições de igualdade nem de desigualdade.
  * Pode até existir, mas a gente aprendeu a como trazer isso para a função objetivo.
* Não tem restrição de igualdade e desigualdade, mas as variavéis de decisão assumem valores binários.
  * Variável Binária: $x,y, z \in \{0,1\}$

## Vamos formular um problema real de investimentos no formato QUBO?

* Queremos montar uma carteira de investimento, onde vamos escolher exatamente 3 ativos entre 5 possiveis.
  * As variáveis assumem valores igual a 0 ou 1, ou seja, não esta na carteira e esta na carteira.
* Queremos escolher ativos que tenham um alto retorno financeiro.
* Queremos baixa covariância entre os ativos escolhidos.
  * A covariância mede a tendência de dois ativos de se moverem na mesma direção ou em direções opostas.





Se eu os seguintes retornos:

\begin{array}{|c|c|}
\hline
\text{Ativo} & \text{Retorno (%)} \\
\hline
x_{1} & 3 \\
x_{2} & 4 \\
x_{3} & 3 \\
x_{4} & 2 \\
x_{5} & 5 \\
\hline
\end{array}

E a seguinte matriz de covariância entre os ativos:

\begin{array}{|c|ccccc|}
\hline
 & x_{1} & x_{2} & x_{3} & x_{4} & x_{5} \\
\hline
x_{1} & 0 & 0.03 & 0.015 & -0.008 & 0.005 \\
x_{2} & 0.03 & 0 & 0.021 & -0.012 & 0.005 \\
x_{3} & 0.015 & 0.021 & 0 & -0.006 & 0.009 \\
x_{4} & -0.008 & -0.012 & -0.006 & 0 & -0.015 \\
x_{5} & 0.005 & 0.005 & 0.009 & -0.015 & 0 \\
\hline
\end{array}


### Dando atenção para o primeiro ponto:

* Queremos montar uma carteira de investimento, onde vamos escolher exatamente 3 ativos entre 5 possiveis.

Temos aqui já a definição das restrições de limite e uma restrição de igualdade.

* A restrição de limite: $x_{1}, x_{2}, x_{3}, x_{4}, x_{5} \in \{0,1\}$
* A restrição de igualdade: $x_{1} + x_{2} + x_{3} + x_{4} + x_{5} = \sum_{i=1}^{5} x_{i} = 3$



### Dando atenção para o segundo ponto:

* Queremos escolher ativos que tenham um alto retorno financeiro.

Temos aqui definido parte da função objetivo:

$Maximizar$ $3x_{1} + 4x_{2} + 3x_{3} + 2x_{4} + 5x_{5} = \sum_{i=1}^{5} retorno_{i}*x_{i} $

Observação: todo problema de maximização pode ser convetido em um de minimização quando se multiplica a função objetivo por -1. O Mesmo vale para transformar o de minimização em maximização.

### Dando atenção para o terceiro ponto:

* Queremos baixa covariância entre os ativos escolhidos.

Temos aqui definido a outra parte da função objetivo:

$Minimizar$ $Cov_{11}*x_{1}*x_{1} + ... + Cov_{54}*x_{5}*x_{4} + Cov_{55}*x_{5}*x_{5} = \sum_{i=1}^{5} \sum_{j=1}^{5} Cov_{ij}*x_{i}*x_{j} $

Vale lembrar aqui que sempre que $i=j$, $Cov_{ij} = 0$.

Se o ativo $i$ foi escolhido e o $j$ não, $Cov_{ij}*x_{i}*x_{j} = 0$. O mesmo vale caso $j$ seja escolhido e $i$ não.

$Cov_{ij}*x_{i}*x_{j}$ só vai ser diferente de zero, representando o valor de $Cov_{ij}$, se forem ativos diferentes e ambos tenham sido escolhido.

### Formulação final

$Minimizar$ $\sum_{i=1}^{5} \sum_{j=1}^{5} Cov_{ij}*x_{i}*x_{j} - \sum_{i=1}^{5} retorno_{i}*x_{i} $

Sujeito a:
* $\sum_{i=1}^{5} x_{i} = 3$
* $x_{1}, x_{2}, x_{3}, x_{4}, x_{5} \in \{0,1\}$



Para deixar na formulação QUBO, basta transformar esta restrição em uma penalização na função objetivo. Logo, temos:

$Minimizar$ $\sum_{i=1}^{5} \sum_{j=1}^{5} Cov_{ij}*x_{i}*x_{j} - \sum_{i=1}^{5} retorno_{i}*x_{i} + \lambda * (\sum_{i=1}^{5} x_{i} - 3)^2$

Sujeito a:
* $x_{1}, x_{2}, x_{3}, x_{4}, x_{5} \in \{0,1\}$

### E agora? Como resolver?

Para questões de demonstração, que tal uma força bruta?

```python
# valores de retorno
retorno = np.array([3, 4, 3, 2, 5])

# Covariâncias
cov = np.array([
    [0, 0.03, 0.015, -0.008, 0.005],
    [0.03, 0, 0.021, -0.012, 0.005],
    [0.015, 0.021, 0, -0.006, 0.009],
    [-0.008, -0.012, -0.006, 0, -0.015],
    [0.005, 0.005, 0.009, -0.015, 0]
])
```

```python
def funcao_objetivo(x):
  value = x @ cov @ x - retorno @ x
  return value

def restricao(x):
  value = np.sum(x) - 3
  return value
```

```python
from itertools import product

# Vetor x
x = [0, 0, 0, 0, 0]

# Gerar todas as combinações possíveis (0 e 1) para o vetor x
combinacoes = np.asarray(list(product([0, 1], repeat=len(x))))
print("Quantidade de combinações: ", len(combinacoes))
```

Equivalente a $2^{n}$, onde $n=5$ neste caso.

```python
import pandas as pd

lamb = 1

# Criar um DataFrame do Pandas com as combinações
df = pd.DataFrame(combinacoes, columns=['x_{}'.format(i+1) for i in range(len(x))])

f_obj = df.apply(funcao_objetivo, axis=1)
res = df.apply(restricao, axis=1)

df["f_obj"] = f_obj
df["res"] = res
df["f_obj_pen"] = df["f_obj"]+ lamb * (df["res"])**2
```
